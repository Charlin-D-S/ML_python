# KNN Complexity


## Step1- Importing required libraries


```python
import matplotlib.pyplot as plt
```

## Step2- Create the data

```python
from sklearn.datasets import make_blobs
X, y = make_blobs(n_samples=50000, centers=2, n_features=2,random_state=0)
```

```python
plt.scatter(X[:, 0],X[:, 1],c=y)
```

## Step3- Preparing the data into train and test

**Split the data into training set and testing set using [train_test_split](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)**

Let's assume we decided to split the dataset into 60% train data and 40% test data.

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
```

## Step4 KNN brute

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3,algorithm='brute')
KNN.fit(X_train,y_train)
```

```python
import datetime
start = datetime.datetime.now()
y_pred = KNN.predict(X_test)
end = datetime.datetime.now()
print(end - start)
```

```python
%%time
y_pred = KNN.predict(X_test)
```

## Step5 KNN ball tree

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3,algorithm='ball_tree')
KNN.fit(X_train,y_train)
```

```python
import datetime
start = datetime.datetime.now()
y_pred = KNN.predict(X_test)
end = datetime.datetime.now()
print(end - start)
```

## Step6 KNN kd tree

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3,algorithm='kd_tree')
KNN.fit(X_train,y_train)
```

```python
import datetime
start = datetime.datetime.now()
y_pred = KNN.predict(X_test)
end = datetime.datetime.now()
print(end - start)
```

## Step7 KNN auto

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors=3,algorithm='auto')
KNN.fit(X_train,y_train)
```

```python
import datetime
start = datetime.datetime.now()
y_pred = KNN.predict(X_test)
end = datetime.datetime.now()
print(end - start)
```
